```python
y =data["Performance_Score"]
X = data[["Years_At_Company","Monthly_Salary","Overtime_Hours","Promotions","Employee_Satisfaction_Score"]]
```

```python
X.columns
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
X_train = scaler.fit_transform(X_train)
```

```python
import joblib
joblib.dump(scaler,"scaler.pkl")
```

```python
X_test = scaler.fit_transform(X_test)
```

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
from sklearn.metrics import accuracy_score
```

```python
def modelperfomance(predictions):
    print("Accuracy score on model is {}".format(accuracy_score(y_test, predictions)))
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
from sklearn.linear_model import LogisticRegression
log_model = LogisticRegression()
```

```python
log_model.fit(X_train , y_train)
```

```python
log_model.predict([[0,0,0,0,0]])
```

```python
log_model.predict(X_test)
```

```python
log_model.predict(X_test)
```

```python
log_predictions = log_model.predict(X_test)
```

```python
modelperfomance(log_predictions)
```

```python
from sklearn.neighbors import KNeighborsClassifier
param_grid = {"n_neighbors": [3,5,7,9,11],
              "weights":["uniform","distance"]}
```

```python
gridkn = GridSearchCV(KNeighborsClassifier(), param_grid,cv = 5)
```

```python
gridkn.fit(X_train,y_train)
```

```python
gridkn.best_params_
```

```python
gridkn.predict(X_test)
```

```python
predictions = gridkn.predict(X_test)
```

```python
modelperfomance(predictions)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import RandomizedSearchCV

param_grid = {
    "n_estimators": [50, 100, 150],
    "max_depth": [5, 10, 20],
    "max_features": ["sqrt", "log2"]
}

random_search = RandomizedSearchCV(
    estimator=RandomForestClassifier(class_weight="balanced"),
    param_distributions=param_grid,
    n_iter=6,
    cv=3,
    n_jobs=-1,
    verbose=2,
    random_state=42
)
```

```python
random_search.fit(X_train, y_train)
```

```python
y_pred = random_search.predict(X_test)
```

```python
modelperfomance(y_pred)
```

```python
joblib.dump(random_search, "model_performance.pkl")
```
